# **Cleaning Data**

```python
# show nulls
df.isnull().sum()
```

```python
# first -----> we can see (ship-date) its not a claen column !
# must change data type to split data in coulmn
# split ( days - months - year ) in single coulmns
```

```python
# for ex : 1101102016
test = df['ship_date'].astype(str)

df['year']=test.str[-4:]
df['dayy']=test.str[-7:-4]
df['day']=df['dayy'].str[:2]
df['months'] =test.str[0:2].astype(int)
df['months_corrected'] = df['months'].apply(lambda x: x if x <= 12 else x//10)

```

```python
df.head(3)
```

```python
# drop coulmns dont need
df.drop(columns=['ship_date','dayy','months'],inplace=True)
```

```python
#df.drop('ship_date_clean',axis=1,inplace=True)
```

```python
# rename coulmn
df.rename(columns={'months_corrected':'month'},inplace=True)
```

```python
df.head(1)
```

```python
# make new ship_date column
df['ship_date'] = pd.to_datetime(df[['year','month','day']],errors='coerce')
```

```python
df.head(1)
```

```python
df.info()
```

```python
df['year']=df['year'].astype(int)
df['day']=df['day'].astype(int)
```

```python
# clean customer_name
df[['first_name', 'last_name']] = df['customer_name'].str.split(' ', n=1, expand=True)
```

```python
df.head(1)
```

```python
df['city']=df['location'].str.split(',',n=1,expand=True)[1]
```

```python
# make total_sales column
df['total_sales']=df['quantity']*df['sales']
```

```python
df['customer_name'].fillna('unknown',inplace=True)
```

# **Machine Learning**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

X = df[['sales', 'profit', 'discount', 'quantity']]
y = LabelEncoder().fit_transform(df['segment'])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

accuracy = accuracy_score(y_test, model.predict(X_test))
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
